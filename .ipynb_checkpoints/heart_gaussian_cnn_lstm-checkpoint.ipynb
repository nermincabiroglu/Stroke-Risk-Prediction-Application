```python
    # import total dataset
    data = pandas.read_csv('D1.csv')
    
       
    # get a list of column names
    headers = list(data.columns.values)
    x = data[headers[:-1]]
    y = data[headers[-1:]].values.ravel()
    
    skf = StratifiedKFold(n_splits=3)

    # blank lists to store predicted values and actual values
    predicted_y = []
    expected_y = []
    predicted_y_train = []
    expected_y_train = []
    # partition data
    for train_index, test_index in skf.split(x, y):
        # specific ".loc" syntax for working with dataframes
        x_train, x_test = x.loc[train_index], x.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # create and fit classifier
        classifier = GaussianNB()
        classifier.fit(x_train, y_train)
        
        # classify our test variables
        predictions_test = classifier.predict(x_test)
        predictions_train = classifier.predict(x_train)
        # save and print accuracy
        accuracy = metrics.accuracy_score(y_test, predictions_test)
        train_accuracy = metrics.accuracy_score(y_train, predictions_train)
        print("Train Accuracy: " + train_accuracy.__str__(),"   Test Accuracy: " + accuracy.__str__())
        
        #print("Train Accuracy: " + train_accuracy.__str__())
        # store result from classification
        predicted_y.extend(classifier.predict(x_test))
        predicted_y_train.extend(classifier.predict(x_train))
        # store expected result for this specific fold
        expected_y.extend(y_test)
        expected_y_train.extend(y_train)
    # save and print accuracy
    accuracytest = metrics.accuracy_score(expected_y, predicted_y)
    accuracytrain= metrics.accuracy_score(expected_y_train, predicted_y_train)
    print("Total Train Accuracy:"+ accuracytrain.__str__(),"   Total Test Accuracy: " + accuracytest.__str__())
```

```python
    # import total dataset
    data = pandas.read_csv('D1.csv')
    
       
    # get a list of column names
    headers = list(data.columns.values)
    x = data[headers[:-1]]
    y = data[headers[-1:]].values.ravel()
    
    skf = StratifiedKFold(n_splits=3)    

    # blank lists to store predicted values and actual values
    predicted_y = []
    expected_y = []
    max_words=50000
    max_len=10
    # partition data
    accSum=0
    val_accSum=0
    sayac=0   #bu ka kere tur olacan belirtir
    
    for train_index, test_index in skf.split(x, y):
        # specific ".loc" syntax for working with dataframes
        x_train, x_test = x.loc[train_index], x.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        sayac=sayac+1
        # create and fit classifier
        model=Sequential()
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(32, activation='relu'))
        model.add(Dense(256, activation='relu', input_shape=(10,)))
        model.add(Dense(32, activation='relu')),
        model.add(Dense(1, activation='sigmoid'))
        model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
        
        #classifier = GaussianNB()
        print(sayac,". Fold")
        
        history=model.fit(x_train, y_train,validation_data= (x_test, y_test), batch_size=64,
                   epochs=300,
                   verbose=2,
                   shuffle= True,
                   class_weight=None,
                   sample_weight=None,
                   initial_epoch=0 )
        history.history.keys()
         #her bir turun epoch larnn deerlerini diziye atar
        print("\n")
        epochDizisi = history.history['accuracy']
        valDizisi = history.history['val_accuracy']
        accSum+= epochDizisi[len(epochDizisi)-1]
        val_accSum+= valDizisi[len(valDizisi)-1]
        print("\n")
     
    print("\nTotal Saya: ",sayac)   
    accMean=accSum/sayac
    val_accMean=val_accSum/sayac
    print("\nTotal Train Accuracy: ",accMean)
    print("\nTotal Test Val_Accuracy: ",val_accMean)
    
```

```python
    # import total dataset
    data = pandas.read_csv('D1.csv')
    
       
    # get a list of column names
    headers = list(data.columns.values)
    x = data[headers[:-1]]
    y = data[headers[-1:]].values.ravel()
    
   
    
    skf = StratifiedKFold(n_splits=3)    

    # blank lists to store predicted values and actual values
    predicted_y = []
    expected_y = []
    max_words=50000
    max_len=10
    # partition data
    accSum=0
    sayac=0   #bu ka kere tur olacan belirtir
    val_accsum=0
    for train_index, test_index in skf.split(x, y):
        # specific ".loc" syntax for working with dataframes
        x_train, x_test = x.loc[train_index], x.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        sayac=sayac+1
        # create and fit classifier
        model=Sequential()
        model.add(Embedding(max_words,50,input_length=max_len))
        model.add(LSTM(64))
        model.add(Dense(256,activation='relu'))
        model.add(Dropout(0.2))
        model.add(Dense(1,activation='sigmoid'))
    
        model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
        
        #classifier = GaussianNB()
        print(sayac,". Fold")
        
        history= model.fit(x_train, y_train,validation_data= (x_test, y_test), batch_size=128,
                   epochs=300,
                   verbose=2,
                   shuffle= True,
                   class_weight=None,
                   sample_weight=None,
                   initial_epoch=0)
        history.history.keys()
         #her bir turun epoch larnn deerlerini diziye atar
        print("\n")
        epochDizisi = history.history['accuracy']
        valDizisi= history.history['val_accuracy']
       # print(history.history['accuracy'])
       # print(history.history['val_accuracy'])
        val_accsum+=valDizisi[len(valDizisi)-1]
        accSum+= epochDizisi[len(epochDizisi)-1]
        print("\n")
    #print(cross_val_score(model,train_index, test_index,scoring='Total Accuracy', cv=3))
    val_accmean=val_accsum/sayac
    accMean=accSum/sayac
    print("\n\nAverage Train Accuracy: ",accMean)
    print("\n\nAverage Validation Accuracy: ",val_accmean)
```
