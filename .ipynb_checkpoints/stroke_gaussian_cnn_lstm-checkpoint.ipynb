```python
max_features = 5000
maxlen = 400
batch_size = 32
embedding_dims = 50
filters = 250
kernel_size = 3
hidden_dims = 250
```

```python
y=data["stroke"]
x=data.drop(["stroke"],axis=1)
#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
#print(len(x_train), 'train sequences')
#print(len(x_test), 'test sequences')
```

```python
# splitting data with cross validation
skf = StratifiedKFold(n_splits=3) 
sayac=0

for train_index, test_index in skf.split(x, y):
    x_train, x_test = x.loc[train_index], x.loc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model = Sequential()
    model.add(Embedding(max_features, # max_features=5000 (it isn't important)
                    embedding_dims, #embedding_dims=50
                    input_length=10)) # length x.shape[0]
                    
    model.add(Dropout(0.2))
    # we add a Convolution1D, which will learn filters
    # word group filters of size filter_length:
    model.add(Conv1D(filters,
                 kernel_size,
                 padding='valid',
                 activation='relu',
                 strides=1))
    # we use max pooling:
    model.add(GlobalMaxPooling1D())

    # We add a vanilla hidden layer:
    model.add(Dense(hidden_dims)) # hidden_dims=250
    model.add(Dropout(0.2))
    model.add(Activation('relu'))

    # We project onto a single unit output layer, and squash it with a sigmoid:
    model.add(Dense(1))
    model.add(Activation('sigmoid'))

    model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

    sayac+=1
    print()
    print((sayac),". Fold")
    print()
    model.fit(x_train, y_train, # train data 
          batch_size=128, # batch_size=32
          epochs=10,
          verbose=2,
          validation_data=(x_test, y_test)) # test data



```

```python
y=data["stroke"]
x=data.drop(["stroke"],axis=1)
_, accuracy = model.evaluate(x, y, batch_size=batch_size, verbose=1)
accuracy
```

```python
    # import total dataset
    data = pd.read_csv('D1.csv')
    
       
    # get a list of column names
    headers = list(data.columns.values)
    x = data[headers[:-1]]
    y = data[headers[-1:]].values.ravel()
    
   
    
    skf = StratifiedKFold(n_splits=3)    

    # blank lists to store predicted values and actual values
    predicted_y = []
    expected_y = []
    max_words=50000
    max_len=10
    # partition data
    accSum=0
    sayac=0   #bu kaç kere tur olacağını belirtir
    val_accsum=0
    for train_index, test_index in skf.split(x, y):
        # specific ".loc" syntax for working with dataframes
        x_train, x_test = x.loc[train_index], x.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        sayac=sayac+1
        # create and fit classifier
        model=Sequential()
        model.add(Embedding(max_words,50,input_length=max_len))
        model.add(LSTM(64))
        model.add(Dense(256,activation='relu'))
        model.add(Dropout(0.2))
        model.add(Dense(1,activation='sigmoid'))
    
        model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
        
        #classifier = GaussianNB()
        print(sayac,". Fold")
        
        history= model.fit(x_train, y_train,validation_data= (x_test, y_test), batch_size=128,
                   epochs=300,
                   verbose=2,
                   shuffle= True,
                   class_weight=None,
                   sample_weight=None,
                   initial_epoch=0)
        history.history.keys()
         #her bir turun epoch larının değerlerini diziye atar
        print("\n")
        epochDizisi = history.history['accuracy']
        valDizisi= history.history['val_accuracy']
       # print(history.history['accuracy'])
       # print(history.history['val_accuracy'])
        val_accsum+=valDizisi[len(valDizisi)-1]
        accSum+= epochDizisi[len(epochDizisi)-1]
        print("\n")
    #print(cross_val_score(model,train_index, test_index,scoring='Total Accuracy', cv=3))
    val_accmean=val_accsum/sayac
    accMean=accSum/sayac
    print("\n\nAverage Train Accuracy: ",accMean)
    print("\n\nAverage Validation Accuracy: ",val_accmean)
```

```python
data = pd.get_dummies(data, columns = ['Residence_type', 'ever_married', 'gender', 'heart_disease', 'hypertension', 'smoking_status', 'work_type'])
standardScaler = StandardScaler()
columns_to_scale = ['age', 'avg_glucose_level', 'bmi']
data[columns_to_scale] = standardScaler.fit_transform(data[columns_to_scale])
```

```python
y = data['stroke']
X = data.drop(['stroke'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
#kneighbours
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    knn_classifier.fit(X_train, y_train)
    knn_scores.append(knn_classifier.score(X_test, y_test))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
knn_edited = np.round(knn_scores [ : ], 2)
for i in range(1,21):
    plt.text(i, knn_edited[i-1], (i, knn_edited[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
print(knn_edited)
```

```python
#Support Vector Classifier
svc_scores = []
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for i in range(len(kernels)):
    svc_classifier = SVC(kernel = kernels[i])
    svc_classifier.fit(X_train, y_train)
    svc_scores.append(svc_classifier.score(X_test, y_test))
svc_edited = np.round(svc_scores [ : ], 2)
colors = rainbow(np.linspace(0, 1, len(kernels)))
plt.bar(kernels, svc_scores, color = colors)
for i in range(len(kernels)):
    plt.text(i, svc_edited[i], svc_edited[i])
plt.xlabel('Kernels')
plt.ylabel('Scores')
plt.title('Support Vector Classifier scores for different kernels')
print(svc_edited)
```

```python
#Decision Tree Classifier
dt_scores = []
for i in range(1, len(X.columns) + 1):
    dt_classifier = DecisionTreeClassifier(max_features = i, random_state = 0)
    dt_classifier.fit(X_train, y_train)
    dt_scores.append(dt_classifier.score(X_test, y_test))
dt_edited = np.round(dt_scores [ : ], 2)
plt.plot([i for i in range(1, len(X.columns) + 1)], dt_scores, color = 'green')
for i in range(1, len(X.columns) + 1):
    plt.text(i, dt_edited[i-1], (i, dt_edited[i-1]))
plt.xticks([i for i in range(1, len(X.columns) + 1)])
plt.xlabel('Max features')
plt.ylabel('Scores')
plt.title('Decision Tree Classifier scores for different number of maximum features')
print(dt_edited)
```

```python
#Random Forest Classifier
rf_scores = []
estimators = [10, 100, 200, 500, 1000]
for i in estimators:
    rf_classifier = RandomForestClassifier(n_estimators = i, random_state = 0)
    rf_classifier.fit(X_train, y_train)
    rf_scores.append(rf_classifier.score(X_test, y_test))
rf_edited = np.round(rf_scores [ : ], 2)
colors = rainbow(np.linspace(0, 1, len(estimators)))
plt.bar([i for i in range(len(estimators))], rf_scores, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_edited[i], rf_edited[i])
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators')
print(rf_edited)
```
